## 一.浏览网页过程

简单来说,浏览网页的过程可分为以下4个步骤：
1. 浏览器通过DNS服务器查找域名对应的IP地址
2. 向IP地址对应的Web服务器发送请求
3. Web服务器响应请求,发回HTML页面
4. 浏览器解析HTML内容,并显示出来
